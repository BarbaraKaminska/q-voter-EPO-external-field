```julia
using CSV, DataFrames 
using DelimitedFiles
using Plots
using Plots.Measures
using Printf
using LaTeXStrings

```

```julia
q = 5
alpha = 0.1
colors_c0 = [:blue, :red]
shapes_c0 = [:circle, :diamond]
ms_ = 1.5
yearvariant = "2018"
# plotname = "EPO_$yearvariant _N$(N)_q$(q)_alpha$(@sprintf("%.2f", alpha))_with_MFA.png"
plotname = "EPO_$(yearvariant)_q$(q)_alpha$(@sprintf("%.2f", alpha))"
# h = [ [0.0, 0.05]]
# h = [[0.01, 0], [0.05, 0], [0, 0.01], [0, 0.05]]
h = [[0, 0], [0.01, 0], [0.05, 0], [0, 0.01], [0, 0.05]]

colors_ = [:black, :red, :green, :blue, :orange, :purple]
lw_ = 2
```

```julia
# pathanalytics = "C:/Users/Basia/OneDrive - Politechnika Wroclawska/Pulpit/PhD/Current work/ExternalInfluence/MFA"
pathanalytics = "qv_EPO_$(yearvariant)"

p1 = plot(xlabel="\$p\$", ylabel="\$c_S\$", xlims=(0, 1), ylims=(0, 1), margin=5mm)
p2 = plot(xlabel="\$p\$", ylabel="\$c_σ\$", xlims=(0, 1), ylims=(0, 1), margin=5mm)
p3 = plot(xlabel="\$p\$", ylabel="\$d\$", xlims=(0, 1), ylims=(0, 1), margin=5mm)

for (idx, htemp) in enumerate(h)
    he, hp = htemp

    try
        file = "EPO_$(yearvariant)_q$(q)_alpha_$(@sprintf("%.2f", alpha))_h_e_$(@sprintf("%.2f", he))_h_p_$(@sprintf("%.2f", hp))_with_stability.txt"
        # Load each file into a DataFrame
        df_ = readdlm(joinpath(pathanalytics, file), '\t')
        df_ = DataFrame(p = df_[:, 1], cE = df_[:, 2] + df_[:, 3], cP = df_[:, 2] + df_[:, 4], d = df_[:, 3] + df_[:, 4], st = df_[:, 6])
       
        df_stable = filter(row -> row.st == 1, df_)
        df_unstable = filter(row -> row.st == 0, df_)

        df_stable = sort(df_stable, [:cE, :cP])  
        df_unstable = sort(df_unstable, [:cE, :cP])


    if length(df_stable.p) == length(unique(df_stable.p))
        
        plot!(p1, df_stable[df_stable.cE .> 0.5, :].p, df_stable[df_stable.cE .> 0.5, :].cE, 
            seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], lw = lw_,
            label = "he = $(@sprintf("%.2f", he)); hp = $(@sprintf("%.2f", hp))")
     
        plot!(p2, df_stable[df_stable.cE .> 0.5, :].p, df_stable[df_stable.cE .> 0.5, :].cP, 
            seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], lw = lw_,
            label = false)  

        plot!(p3, df_stable[df_stable.cE .> 0.5, :].p, df_stable[df_stable.cE .> 0.5, :].d, 
            seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], lw = lw_,
            label = false)    

    
    else 
        print("Double hysteresis")
        duplicates = df_stable.p[findall(x -> count(==(x), df_stable.p) > 1, df_stable.p)]
        max_p = maximum(duplicates)
        # println(" Max p with double hysteresis: $(@sprintf("%.2f", max_p))")
        c_up = maximum(df_stable.cE[df_stable.p .== max_p])
        println(" Max p with double hysteresis: $(@sprintf("%.2f", max_p)), cE at that point: $(@sprintf("%.2f", c_up))")

        df_temp = df_stable[df_stable.cE .> 0.5, :]

        plot!(p1, df_temp[df_temp.cE .> c_up, :].p, df_temp[df_temp.cE .> c_up, :].cE, 
            seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], lw = lw_,
            label = "he = $(@sprintf("%.2f", he)); hp = $(@sprintf("%.2f", hp))")

        plot!(p2, df_temp[df_temp.cE .> c_up, :].p, df_temp[df_temp.cE .> c_up, :].cP, 
            seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], lw = lw_,
            label = false)  

        plot!(p3, df_temp[df_temp.cE .> c_up, :].p, df_temp[df_temp.cE .> c_up, :].d, 
            seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], lw = lw_,
            label = false)    

        plot!(p1, df_temp[df_temp.cE .< c_up, :].p, df_temp[df_temp.cE .< c_up, :].cE, 
            seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], lw = lw_,
            label = false)

        plot!(p2, df_temp[df_temp.cE .< c_up, :].p, df_temp[df_temp.cE .< c_up, :].cP, 
            seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], lw = lw_,
            label = false)  

        plot!(p3, df_temp[df_temp.cE .< c_up, :].p, df_temp[df_temp.cE .< c_up, :].d, 
            seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], lw = lw_,
            label = false)    

    end

    plot!(p1, df_unstable[df_unstable.cE .> 0.5, :].p, df_unstable[df_unstable.cE .> 0.5, :].cE, 
        seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], ls = :dash, lw = lw_,
        label = false)  

    plot!(p2, df_unstable[df_unstable.cE .> 0.5, :].p, df_unstable[df_unstable.cE .> 0.5, :].cP, 
        seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], ls = :dash, lw = lw_,
        label = false)   

    plot!(p3, df_unstable[df_unstable.cE .> 0.5, :].p, df_unstable[df_unstable.cE .> 0.5, :].d, 
        seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], ls = :dash, lw = lw_,
        label = false)  

    plot!(p1, df_stable[df_stable.cE .< 0.5, :].p, df_stable[df_stable.cE .< 0.5, :].cE, 
        seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], lw = lw_,
        label = false)        
    plot!(p1, df_stable[df_stable.cE .== 0.5, :].p, df_stable[df_stable.cE .== 0.5, :].cE, 
        seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], lw = lw_,
        label = false)


    plot!(p2, df_stable[df_stable.cE .< 0.5, :].p, df_stable[df_stable.cE .< 0.5, :].cP, 
        seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], lw = lw_,
        label = false)
    plot!(p2, df_stable[df_stable.cE .== 0.5, :].p, df_stable[df_stable.cE .== 0.5, :].cP, 
        seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], lw = lw_,
        label = false)


    plot!(p3, df_stable[df_stable.cE .< 0.5, :].p, df_stable[df_stable.cE .< 0.5, :].d, 
        seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], lw = lw_,
        label = false)    
    plot!(p3, df_stable[df_stable.cE .== 0.5, :].p, df_stable[df_stable.cE .== 0.5, :].d, 
        seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], lw = lw_,
        label = false)


    plot!(p1, df_unstable[df_unstable.cE .< 0.5, :].p, df_unstable[df_unstable.cE .< 0.5, :].cE, 
        seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], ls = :dash, lw = lw_,
        label = false)      
    plot!(p1, df_unstable[df_unstable.cE .== 0.5, :].p, df_unstable[df_unstable.cE .== 0.5, :].cE, 
        seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], ls = :dash, lw = lw_,
        label = false)
    

    plot!(p2, df_unstable[df_unstable.cE .< 0.5, :].p, df_unstable[df_unstable.cE .< 0.5, :].cP, 
        seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], ls = :dash, lw = lw_,
        label = false)     
    plot!(p2, df_unstable[df_unstable.cE .== 0.5, :].p, df_unstable[df_unstable.cE .== 0.5, :].cP, 
        seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], ls = :dash, lw = lw_,
        label = false)


    plot!(p3, df_unstable[df_unstable.cE .< 0.5, :].p, df_unstable[df_unstable.cE .< 0.5, :].d, 
        seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], ls = :dash, lw = lw_,
        label = false)   
    plot!(p3, df_unstable[df_unstable.cE .== 0.5, :].p, df_unstable[df_unstable.cE .== 0.5, :].d, 
        seriestype=:line, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, color = colors_[idx], ls = :dash, lw = lw_,
        label = false)



        # plot!(p1, df_stable.p, df_stable.cE, seriestype=:scatter, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14,
        # label = "he = $(@sprintf("%.2f", he)); hp = $(@sprintf("%.2f", hp))")
        # plot!(p2, df_stable.p, df_stable.cP, seriestype=:scatter, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, label = false)
        # plot!(p3, df_stable.p, df_stable.d, seriestype=:scatter, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, label = false)

        # plot!(p1, df_unstable.p, df_unstable.cE, seriestype=:scatter, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14,
        # label = "he = $(@sprintf("%.2f", he)); hp = $(@sprintf("%.2f", hp))")
        # plot!(p2, df_unstable.p, df_unstable.cP, seriestype=:scatter, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, label = false)
        # plot!(p3, df_unstable.p, df_unstable.d, seriestype=:scatter, framestyle=:box, ms = ms_, msw = 0.1, xlabelfontsize=14, ylabelfontsize=14, label = false)


    catch e
        println(e)
        continue
    end
end

plot(p1, p2, p3, layout = (1, 3), size=(1200, 400), 
     plot_title="Model $yearvariant, q = $q, α = $(@sprintf("%.2f", alpha))", 
     plot_titlefontsize=14, dpi=300)


plotname_suffix = "full_with_stability.png"

savefig(joinpath("qv_EPO_figures_$(yearvariant)/", plotname*plotname_suffix))
```
